```python
import os
import cv2
from ultralytics import YOLO

# Define the directories
input_dir = 'D:\\Projects Computer Vision\\Computer vision Project 3 Skin-Disease-Detection\\Skin-Disease-Detection github repo (bad)\\Skin-Disease-Detection\\dataset_2\\train'
output_dir = 'D:\\Projects Computer Vision\\Computer vision Project 3 Skin-Disease-Detection\\output_images-best(acne-1)(200).pt'

# Load the trained YOLO model
model = YOLO('Models (Yolo models)/best(acne-1)(200).pt')

# Define a function to draw bounding boxes without labels
def draw_bounding_boxes(image, boxes, color=(255,215,0)):
    """
    Draws bounding boxes on the image without labels and with a specified color.
    
    :param image: The image on which to draw the bounding boxes.
    :param boxes: A list of bounding boxes, where each box is represented as [x_min, y_min, x_max, y_max].
    :param color: A tuple representing the color of the bounding box (B, G, R).
    """
    for box in boxes:
        x_min, y_min, x_max, y_max = map(int, box[:4])
        
        # Draw rectangle
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)
        
    return image

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Iterate over all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):  # Add more extensions if needed
        # Full path of the image
        image_path = os.path.join(input_dir, filename)
        
        # Perform prediction on the image
        results = model.predict(image_path)
        
        # Print the results
        print(f"Results for {filename}:")
        print(results[0])
        print("==========")
        
        # Extract the bounding boxes
        boxes = results[0].boxes.xyxy.cpu().numpy()  # Convert tensor to numpy array
        
        # Read the image
        image = cv2.imread(image_path)
        
        # Draw the bounding boxes without labels
        image_with_boxes = draw_bounding_boxes(image, boxes)
        
        # Save the result image to the output directory
        result_image_path = os.path.join(output_dir, filename)
        cv2.imwrite(result_image_path, image_with_boxes)

print("Predictions completed and saved.")
```
